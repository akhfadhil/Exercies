````python
import requests
import lxml
import bs4
````

````python
base_url = 'http://quotes.toscrape.com/page/{}/'
res = requests.get(base_url.format(1))
````

````python
res.text
````

````python
soup = bs4.BeautifulSoup(res.text, 'lxml')
soup
````

author

````python
author = []
authors = soup.select('.quote')
for container in authors:
  span = container.select('span')[1]
  author.append(span.select('small')[0].text)
# for auth in authors:
#   author.append(auth.text)
#   # if auth.text not in author:
#   #   author.append(auth.text)
author = set(author)
````

````python
author
````

quote

````python
quotes = []

for quote in soup.select('.quote'):
  quotes.append(quote.select('span')[0].text)
````

````python
quotes
````

tag

````python
top10tags = []
for i in range(len(soup.select('.tag-item'))):
  top10tags.append(soup.select('.tag-item')[i].select('.tag')[0].text)
top10tags
````

author all pages

````python
base_url = 'http://quotes.toscrape.com'
page = 1

authAll = []

while True:
  page_url = base_url + '/page/{}/'.format(page)
  res = requests.get(page_url)
  soup = bs4.BeautifulSoup(res.text, 'lxml')

  if 'No quotes found!' in str(soup):
    break

  authPage = soup.select('small')
  for i in range(len(authPage)):
    authAll.append(authPage[i].text)

  page = page + 1

authAll = set(authAll)
````
